```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on="title")
```

 # DATA TO KEEP
 #   1. Generes
 #   2. Id
 #   3. Keywords
 #   4. Title
 #   5. Overview
 #   6. Cast
 #   7.Crew

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

# Convert
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# TO
# ['Action', 'Adventure', 'Fantasy', 'Scifi']

```python
import ast
def convert(obj):
    res = []
    for i in ast.literal_eval(obj):
        res.append(i['name'])
    return res
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convertFirst3(obj):
    res = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            res.append(i['name'])
        else:
            break
        counter+=1
    return res
```

```python
movies['cast'] = movies['cast'].apply(convertFirst3)
```

```python
def fetch_director(obj):
    res = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            res.append(i['name'])
            break
    return res
```

```python
movies['director'] = movies['crew'].apply(fetch_director)
movies=movies.drop(['crew'], axis=1)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(5)
```

```python
# remove spaces
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['director']=movies['director'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(5)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['director']
```

```python
movies.head(5)
```

```python
data = movies[['movie_id', 'title', 'tags']]
```

```python
data['tags']=data['tags'].apply(lambda x:" ".join(x))
```

```python
data['tags']=data['tags'].apply(lambda x:x.lower())
```

```python
data.head(5)
```

# How to recommend?
## We will use vectorization, and we will acheive it using 'bag of words' method 

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
```

```python
cv.get_feature_names()
```

# Apply Steming to make similar words into a singleform
## Eg: ['actor', 'actors'] to ['actor', 'actor']

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y);
```

```python
data['tags'] = data['tags'].apply(stem)
```

```python
data.head(5)
```

```python
vectors = cv.fit_transform(data['tags']).toarray()
cv.get_feature_names()
```

# Now that we have created 5000 dimensional vector for all movies, calculate diistance between them.
# Calculate cosine distance not Euclidean Distance, because dimensionality is High

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    res=[]
    movie_index = data[data['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        res.append(data.iloc[i[0]].title)
        
    return res
```

```python
recommend('The Dark Knight Rises')
```
